```python
sql = """DESCRIBE daily_price"""
pd.read_sql(sql, con=conn)
```

```python
# get ticker list
sql = """SELECT ticker, id FROM security"""
tickers = pd.read_sql(sql, con=conn)
```

```python
tickers.shape
```

```python
# missing values for these tickers 
tickers_noData = []
 # populating to daily_price
for row in tickers.itertuples():
    ticker = row.ticker
    ticker_id = row.id
#     print(ticker)
    try:
        tickers_price = price_df[ticker]
        tickers_price = tickers_price[tickers_price['Open'].notnull()]
        tickers_price.reset_index(inplace=True)
        tickers_price['Date'] = tickers_price['Date'].astype(str)
        for row in tickers_price.itertuples(index=False):
            # assign yahoo finance as 9
            values = [9, ticker_id]+ list(row)
            cursor.execute("""INSERT INTO daily_price (data_vendor_id, 
            ticker_id, price_date, open_price, high_price, low_price, 
            close_price, adj_close_price, volume) VALUES 
            (%s, %s, %s, %s, %s, %s, %s, %s, %s)""",
            tuple(values))
#             print(tuple(values))
    except:
        print(ticker)
        tickers_noData.append(ticker)
        pass
conn.commit()
```
